## Ballon D'or Prediction

### I will collect data from the top 5 leagues in Europe, starting with player stats from each season from the past 20 years (00/01 to 20/21)

```python
import requests
from bs4 import BeautifulSoup

import csv
import sys
from os import path

import numpy as np
import pandas as pd
```

### Player Standings

```python
url = 'https://statbunker.com/competitions/PlayerStandings'
```

```python
def fetch_competition(competition):
    print(f"scraping competition with id {competition}")
    response = requests.get(
        f"{url}?comp_id={competition}"
    )
    soup = BeautifulSoup(response.content, "html.parser")
    return soup
```

```python
def add_players_stats(soup, dict):
    for players_html in soup.find_all("tr", title = True):
        dict['Players'].append(players_html.find("td", class_ = "mob tdInline").string)
        dict['Clubs'].append(players_html.find_all("td", class_ = "nonMob", title = None)[0].string)
        dict['Position'].append(players_html.find_all("td", class_ = "nonMob", title = None)[1].string)
        dict['Total'].append(players_html.find_all("td", class_ = "nonMob", title = "")[2].string)
        dict['Goals'].append(players_html.find("td", class_ = "mob", title = "Goals").string)
        dict['A'].append(players_html.find("td", class_ = "mob", title = "Assists").string)
        dict['Start'].append(players_html.find("td",class_ = "nonMob", title = "Total started").string)
        dict['Sub'].append(players_html.find("td",class_ = "nonMob", title = "Started as a sub").string)
        dict['CO'].append(players_html.find("td",class_ = "nonMob", title = "Came on").string)
        dict['Off'].append(players_html.find("td",class_ = "nonMob", title = "Taken off").string)
        dict['Pen S'].append(players_html.find("td",class_ = "nonMob", title = "Penalty Scored").string)
        dict['Pen SV'].append(players_html.find("td",class_ = "nonMob", title = "Penalty Saved").string)
        dict['Pen M'].append(players_html.find("td",class_ = "nonMob", title = "Penalty Missed").string)
        dict['Pen C'].append(players_html.find("td",class_ = "nonMob", title = "Pen conceded").string)
        dict['OG'].append(players_html.find("td",class_ = "nonMob", title = "Own goals").string)
        dict['League'].append(soup.find("title").string[26:-6])
        dict['Season'].append(soup.find("title").string[-6:])
```

```python
dic = {'Players': [],
       'Clubs': [],
       'Position': [],
       'Total': [],
       'Goals': [],
       'A': [],
       'Start': [],
       'Sub': [],
       'CO': [],
       'Off': [],
       'Pen S': [],
       'Pen SV': [],
       'Pen M': [],
       'Pen C': [],
       'OG': [],
       'League': [],
       'Season': []
      }
```

```python
def create_players_df(max_page):
    for page in range(max_page):
        soup = fetch_competition(page)
        add_players_stats(soup, dic)
    return pd.DataFrame.from_dict(dic, orient = 'index').transpose()
```

```python
scraping_df = create_players_df(710)
```
